```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#load dataset
data = np.loadtxt("data.txt")
```

```python
def tde(x, t):
    """calculate time delay embedding

    Args:
        x (list): data
        t (int): delay time

    Returns:
        list: TDE of data
    """
    y = np.zeros(len(x))
    for i in range(len(x)-t):
        y[i] = x[i+t]
    return y
```

```python
T10 = tde(data, 10)
T100 = tde(data, 100)
T1000 = tde(data, 1000)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8,10))

axs[0].plot(data, T10,".", color = "red")
axs[0].set_xlabel('X(t)')
axs[0].set_ylabel(r'$\tau = 10$')

axs[1].plot(data, T100, ".", color = "blue")
axs[1].set_xlabel('X(t)')
axs[1].set_ylabel(r'$\tau = 100$')

axs[2].plot(data, T1000, ".", color = "magenta")
axs[2].set_xlabel('X(t)')
axs[2].set_ylabel(r'$\tau = 1000$')
plt.tight_layout()

plt.savefig("TDE")
```
